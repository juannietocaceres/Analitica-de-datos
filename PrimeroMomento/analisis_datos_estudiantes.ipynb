Primero, se importaron las librerías necesarias: pandas para la manipulación de datos, matplotlib y seaborn para la visualización. Luego, se cargó el archivo CSV mediante la función read_csv, almacenándolo en un DataFrame llamado df.

Posteriormente, se utilizó el método head() para visualizar las primeras filas del dataset. Esto permite tener una idea general de cómo están organizados los datos, qué variables contiene y cómo se distribuyen los valores.

## 📅 Semana 3 – Limpieza de datos

En esta etapa se realizó la verificación de la calidad de los datos para asegurar que el análisis posterior no esté afectado por errores o inconsistencias.

Primero, se utilizó el método isnull().sum() para identificar la presencia de valores nulos en cada una de las columnas.
Los resultados muestran que todas las variables tienen un valor de 0, lo que indica que no existen datos faltantes en el dataset.

Posteriormente, se aplicaron dos procesos de limpieza:

dropna() para eliminar posibles valores nulos

drop_duplicates() para eliminar registros duplicados

Aunque en este caso no se encontraron valores nulos, estas operaciones se realizaron como una buena práctica dentro del proceso de análisis de datos.

Esto garantiza que el dataset esté completamente limpio, evitando sesgos o errores en las etapas siguientes del análisis.

```python
df.isnull().sum()
```

```python
df = df.dropna()
df = df.drop_duplicates()
```

Se verificó que los datos estén limpios para evitar sesgos en el análisis.

## 📅 Semana 4 – Análisis exploratorio

En esta etapa se realizó un análisis exploratorio de los datos con el objetivo de identificar patrones, relaciones y comportamientos dentro del dataset.

## 📊 1. Distribución por género

Se utilizó value_counts() para analizar la cantidad de estudiantes por género.

Los resultados muestran:

Masculino: 753

Femenino: 747

Esto indica que el dataset está balanceado, ya que ambos géneros tienen una representación muy similar, lo cual es positivo para el análisis.





```python
df['genero'].value_counts()
```

## 📊 2. Puntaje promedio por estrato

Se utilizó groupby() para calcular el promedio del puntaje global según el estrato socioeconómico.

Se observa que:

Los puntajes son relativamente similares entre estratos

No hay una diferencia extremadamente marcada

Sin embargo, existen pequeñas variaciones que pueden indicar cierta influencia del contexto socioeconómico

Esto sugiere que el estrato podría tener un impacto moderado en el rendimiento académico.

```python
df.groupby('estrato')['puntaje_global'].mean()
```

## 📊 3. Puntaje promedio por zona

Se compararon los resultados entre zona rural y urbana.

Resultados:

Rural: ~276.5

Urbana: ~274.9

Esto muestra que no hay una diferencia significativa entre zonas, lo que indica que el rendimiento académico es bastante similar en ambos contextos.



```python
df.groupby('zona')['puntaje_global'].mean()
```

## 📊 4. Análisis de correlación

Se utilizó corr() para identificar relaciones entre variables numéricas.

Hallazgos importantes:

Existe una correlación positiva fuerte entre:

puntaje_global y:

matemáticas (~0.59)

lectura (~0.56)

ciencias (~0.58)

👉 Esto tiene sentido, ya que el puntaje global depende de estas variables.

La edad tiene correlación muy baja → no influye significativamente

El estrato tiene correlación casi nula → baja influencia directa

```python
df.corr(numeric_only=True)
```

Se analizaron relaciones importantes como el impacto del estrato y la zona en el rendimiento académico.

## 📅 Semana 5 – Visualización

## 📊 1. Gráfico de barras – Distribución por género

Se utilizó countplot para visualizar la cantidad de estudiantes por género.

El gráfico muestra que:

La cantidad de estudiantes masculinos y femeninos es muy similar

Esto confirma que el dataset está balanceado

Este tipo de gráfico permite comparar categorías de manera clara y rápida.

```python
sns.countplot(x='genero', data=df)
plt.show()
```

## 📊 2. Histograma – Distribución del puntaje global

Se utilizó histplot para analizar cómo se distribuyen los puntajes globales.

Se observa que:

Los puntajes se concentran en un rango medio

No hay valores extremadamente altos o bajos

La distribución es relativamente uniforme

Esto permite entender el comportamiento general del rendimiento académico.

```python
sns.histplot(df['puntaje_global'])
plt.show()
```

## 📊 3. Boxplot – Comparación por género

Este gráfico permite analizar:

La mediana de los puntajes

La dispersión de los datos

Posibles valores atípicos (outliers)

Permite comparar si existe diferencia en el rendimiento entre hombres y mujeres.

```python
sns.boxplot(x='zona', y='puntaje_global', data=df)
plt.show()
```

## 📊 4. Boxplot – Estrato vs puntaje

Este gráfico muestra:

Cómo varía el puntaje según el estrato

Si hay diferencias entre niveles socioeconómicos

Se puede observar si el contexto social influye en el rendimiento.

```python
sns.boxplot(x='estrato', y='puntaje_global', data=df)
plt.show()
```

## 📊 5. Mapa de calor – Correlaciones

El heatmap permite visualizar la relación entre variables:

Colores más intensos → mayor relación

Colores claros → menor relación

Se confirma que:

Los puntajes por materia están relacionados con el puntaje global

Variables como edad o estrato tienen menor impacto

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

## 📊 6. Gráfico de dispersión – Edad vs puntaje

Este gráfico permite ver si existe relación entre la edad y el rendimiento.

Se observa que:

No hay una relación clara

Los datos están dispersos

Esto indica que la edad no influye significativamente en el puntaje.

```python
sns.scatterplot(x='edad', y='puntaje_global', data=df)
plt.show()
```

Se utilizaron gráficos para identificar tendencias, diferencias entre grupos y correlaciones.
